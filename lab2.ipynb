**Best-First Search 8-puzzle solver**

```python
import heapq

# Board size
n = 3

goal = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 0]
]

# Moves: Down, Up, Right, Left
row = [1, -1, 0, 0]
col = [0, 0, 1, -1]

class State:
    def __init__(self, board, x, y, h):
        self.board = board
        self.x = x
        self.y = y
        self.h = h

    def __lt__(self, other):
        return self.h < other.h

def heuristic(board):
    count = 0
    for i in range(n):
        for j in range(n):
            if board[i][j] != 0 and board[i][j] != goal[i][j]:
                count += 1
    return count

def is_valid(x, y):
    return 0 <= x < n and 0 <= y < n

def best_first_search(start, x, y):
    pq = []
    visited = set()
    h = heuristic(start)
    heapq.heappush(pq, State(start, x, y, h))
    visited.add(tuple(map(tuple, start)))
    step = 0

    while pq:
        current = heapq.heappop(pq)
        step += 1
        print(f"\nStep {step} | Heuristic = {current.h}")
        for r in current.board:
            print(r)

        if current.board == goal:
            print("\n GOAL FOUND!")
            return

        for i in range(4):
            new_x = current.x + row[i]
            new_y = current.y + col[i]

            if is_valid(new_x, new_y):
                new_board = [r[:] for r in current.board]
                new_board[current.x][current.y], new_board[new_x][new_y] = \
                new_board[new_x][new_y], new_board[current.x][current.y]

                board_tuple = tuple(map(tuple, new_board))

                if board_tuple not in visited:
                    h = heuristic(new_board)
                    heapq.heappush(pq, State(new_board, new_x, new_y, h))
                    visited.add(board_tuple)

    print(" No solution found")

# Initial state
start = [
    [1, 2, 3],
    [4, 6, 0],
    [7, 5, 8]
]

best_first_search(start, 1, 2)
```
